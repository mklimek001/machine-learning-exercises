### European footbal players market value predictions

Data from https://www.kaggle.com/datasets/alessiasimone/european-soccer-dataset-season-20172020

##### Data preparation

```python
import numpy as np
import pandas as pd
import sklearn
sklearn.__version__
```

```python
raw_df = pd.read_csv(r'/content/european_football.csv', sep=';')
raw_df.head()
```

```python
raw_df.columns
```

```python
raw_df.info()
```

```python
raw_df.shape
```

##### Deleting some columns

```python
df = raw_df.drop(['Season', 'WinCL',  'LgRk', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'Pts',
                  'Attendance','games starts', 'passes completed', 'sca', 'gca', 'dribbles completed',
                  'pens won', 'goals against gk', 'shots on target against',
                  'CLBestScorer', 'crosses', 'tackles', 'pressures', 'blocks', 'squad', 'Season'], axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop('player', inplace=True, axis=1)
df
```

```python
df.drop('nationality', inplace=True, axis=1)
df
```

```python
df.info()
```

```python
categorical_cols = [column for column in df.columns if df[column].dtype == 'object']
categorical_cols
```

```python
for col in categorical_cols:
  df[col] = df[col].astype('category')

df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0, inplace=True)
df.isnull().sum()
```

```python
numerical_cols = [column for column in df.columns if df[column].dtype in ['int64', 'float64']]
numerical_cols
```

```python
df = pd.get_dummies(df, drop_first=True)
df
```

##### Scaling numerical data

```python
numerical_cols.remove('value')

for col in numerical_cols:
  df[col] = (df[col] - df[col].min())/(df[col].max() - df[col].min())

df
```

```python
X = df.copy()
```

```python
y = X.pop('value')
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=23)

print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape) 
```

##### Linear regression

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
predictions = pd.DataFrame(data={
    'y_true': y_test,
    'y_pred': y_pred
    })

predictions['error'] = predictions['y_true'] - predictions['y_pred']
predictions
```

```python
from sklearn.metrics import max_error, mean_absolute_error, mean_squared_error

print('R2: ', regressor.score(X_test, y_test))
print('MAX: ', max_error(y_test, y_pred))
print('MAE: ', mean_absolute_error(y_test, y_pred))
print('MSE: ', mean_squared_error(y_test, y_pred)) 
```

##### Decison tree

```python
from sklearn.tree import DecisionTreeRegressor

dt_regressor = DecisionTreeRegressor(max_depth=1)
dt_regressor.fit(X_train, y_train)
```

```python
print('R2: ', dt_regressor.score(X_test, y_test))
```

```python
dt_regressor = DecisionTreeRegressor(max_depth=5)
dt_regressor.fit(X_train, y_train)
print('R2: ', dt_regressor.score(X_test, y_test))
```

```python
dt_regressor = DecisionTreeRegressor(max_depth=10)
dt_regressor.fit(X_train, y_train)
print('R2: ', dt_regressor.score(X_test, y_test))
```

```python
dt_regressor = DecisionTreeRegressor(max_depth=100)
dt_regressor.fit(X_train, y_train)
print('R2: ', dt_regressor.score(X_test, y_test))
```

```python
best_dt_regressor = None
best_r2 = 0
depth = 0

for i in range(1, 100):
  curr_dt_regressor = DecisionTreeRegressor(max_depth=i)
  curr_dt_regressor.fit(X_train, y_train)
  curr_r2 = curr_dt_regressor.score(X_test, y_test)
  if curr_r2 > best_r2:
    best_r2 = curr_r2
    depth = i
    best_dt_regressor = curr_dt_regressor


print("Best R2 score: ", best_r2)
print("Depth param: ", depth)
```

```python
y_pred_dt = best_dt_regressor.predict(X_test)
```

```python
predictions['y_pred_dt'] = y_pred_dt
predictions['error_dt'] = predictions['y_true'] - predictions['y_pred_dt']
predictions
```
